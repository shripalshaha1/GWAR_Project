# Data Transformation

## Data Standardization

Out of standard range Columns: duration_ms, loudness, key, popularity and tempo

We can fix the data by applying StandardScaler scaling to each column and fixing the range in a more standard range

```python
from sklearn.preprocessing import StandardScaler
```

```python
df[['duration_ms', 'loudness', 'key', 'popularity', 'tempo']].describe()
```

```python
scaler = StandardScaler()
df[['duration_ms', 'loudness', 'key', 'popularity', 'tempo']] = scaler.fit_transform(df[['duration_ms', 'loudness', 'key', 'popularity', 'tempo']])
```

```python
df[['duration_ms', 'loudness', 'key', 'popularity', 'tempo']].describe()
```

```python
df.head()
```

## Data Regularization

Apply Lasso(L1) Regularization to Extract important features

```python
X = df.drop('popularity', axis=1)
y = df['popularity']
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lasso = SelectFromModel(Lasso(alpha=0.01))
lasso.fit(X, y)
```

```python
features = list(X.columns[lasso.get_support()])
features.append('popularity')
features
```

```python
df = df[features]
df.head()
```

## Data Reduction Using PCA

Apply PCA for dimentionality reduction and reduce the number of components to 3.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pc = pca.fit_transform(df)
pc
```

```python
df[['c0', 'c1', 'c2']] = pc
```

Recommendation Generation

```python
X['label'] = label
```

```python
indices = pd.Series(data.index, index=data['name']).drop_duplicates()

def recommend(song_name, amount=10):
  index = indices[song_name]

  song = X.loc[index[0]]
  rec = X.drop(index[0])

  distance = []
  for songs in rec.values:
      d = 0
      for col in np.arange(len(rec.columns)):
          d = d + np.absolute(float(song[col]) - float(songs[col]))
      distance.append(d)

  rec['distance'] = distance
  rec = rec.sort_values('distance')

  idx = rec.iloc[:amount].index

  print(indices[idx])
```

```python
recommend("Thunderstruck")
```
